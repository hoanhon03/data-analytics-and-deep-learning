3. Vấn đề 1: Tiến hành tải dữ liệu vào chương trình ứng dụng Python và giải quyết vấn đề “Missing header in the csv file”.

```python
import pandas as pd
from pandas import ExcelFile

pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
pd.set_option('display.width', None)    

# Problem 1: thêm header vào datafram để diển giải dữ liệu 
column_names = ["Id", "Name", "Age", "Weight", "m0006", "m0612", "m1218", "f0006", "f0612", "f1218"]
# đọc file
F=pd.read_csv("patient_heart_rate.csv", names=column_names, on_bad_lines='warn')
# hiển thị 
F.head()
```

Vấn đề 2: Xử lý vấn đề một cột lưu hỗn hợp nhiều dữ liệu, ở đây là cột “Name” chứa bao 
gồm “Firstname” và “Lastname”, giải pháp là ta sẽ tách ra làm 2 cột

```python
F[['Firstname', 'Lastname']]=F['Name'].str.split(expand=True)
data_splitname=F.drop("Name", axis=1)
data_splitname
```

Vấn đề 3: Cột Weight có vấn đề về không thống nhất các đơn vị đo lường trong dữ liệu. Ta sẽ chuyển các đơn vị về thành đơn vị chuẩn “kg”

```python
weight=F['Weight']
for i in range(0, len(weight)):
    x=str(F.loc[i, 'Weight'])
    if "lbs" in x[-3:]:
        # Remove the 'lbs' from the value
        x = x[:-3]
        # Convert string to float
        float_x = float(x)
        # Convert to kgs and store as int
        y = int(float_x / 2.2)
        # Convert back to string
        y = str(y) + "kgs"
        F.loc[i, 'Weight'] = y # .loc => tác động trực tiếp đến dataframe gốc
print(F)
```

Vấn đề 4: Vấn đề về xuất hiện dòng dữ liệu rỗng (không có giá trị: NaN). Giải pháp có thể đưa ra là xóa bỏ

```python
# Problem 4: 
F.dropna(how="all", inplace=True)
print(F)
```

Vấn đề 5: Có nhiều dòng dữ liệu bị trùng lắp thông tin hoàn toàn[fullname, lastname, 
age, weight,....], giải pháp đưa ra là chỉ giữ lại một dòng dữ liệu, tuy nhiên giải pháp phải dựa trên nghiệp vụ của tập dữ liệu và quan sát của người xử lý.

```python
# Problem 5: 
F = F.drop_duplicates(subset=['Firstname', 'Lastname', 'Age', 'Weight'])
print(F)
```

Vấn đề 6: Xuất hiện dữ liệu bị ảnh hưởng bởi lỗi non-ASCII, không định dạng ASCII. 
Giải pháp: Tùy vào nghiệp vụ ta có thể: xóa dữ liệu tại đó, thay thế bằng dữ liệu khác hoặc thay bằng việc đánh dấu bằng một kí tự khác (ví dụ: ‘warning’)

```python
# Problem 6:
F['Firstname'] = F['Firstname'].replace({r'[^\x00-\x7F]+': ''}, regex=True)
F['Lastname'] = F['Lastname'].replace({r'[^\x00-\x7F]+': ''}, regex=True)
print(F)
```

Vấn đề 7: “Missing values”, vấn đề này xảy ra tại các cột “Age”, “Weight” và “Heart 
Rate”. Thiếu dữ liệu (dữ liệu không đầy đủ) là vấn đề xảy ra nhiều trong các nguồn dữ liệu do nhiều nguyên nhân chủ quan lẫn khách quan.Có một vài giải pháp để xử lý vấn đề này, chủ yếu dựa trên kinh nghiệm và nghiệp vụ về tập dữ liệu đó. Một số giải pháp đưa đề xuất từ chuyên gia như sau:
    a. Deletion: Remove records with missing values
    b. Dummy substitution: Replace missing values with a dummy but valid value: 
            e.g.: 0 for numerical values.
    c. Mean substitution: Replace the missing values with the mean.
    d. Frequent substitution: Replace the missing values with the most frequent item.
    e. Improve the data collector: Your business folk will talk to the clients and inform them about why it is worth fixing the problem with the data collector.
Yêu cầu:
- Thống kê thông tin dữ liệu thiếu trên từng biến Age và Weight 
- Yêu cầu xử lý dữ liệu thiếu như sau: Nếu dòng nào có Age hoặc Weight có dữ liệu thì phần Age hoặc Weight được tính như bên dưới, nếu thiếu cả 2 thông tin thì xóa dòng
    o Age: Giá trị thay thế là mean của các giá trị trong cột Age

```python
#problem 7:
# Thống kê thông tin dữ liệu thiếu
missing_age = F['Age'].isnull().sum()
missing_weight = F['Weight'].isnull().sum()
print(f"Số lượng giá trị thiếu trong cột Age: {missing_age}")
print(f"Số lượng giá trị thiếu trong cột Weight: {missing_weight}")

# Tính giá trị trung bình của cột Age
mean_age = round(F['Age'].mean())

# Xử lý dữ liệu thiếu
F['Age'].fillna(mean_age, inplace=True)
F.dropna(subset=['Age', 'Weight'], how='all', inplace=True)
F


```

Vấn đề 8: “một cột chứa quá nhiều thông tin cần được phân rã”, như trong bài toán này ta thấy header “m0006” chứa các nội dung bao gồm: m → male, 1218 ~ 12-18 (mm-dd). Còn giá trị thì là kết quả huyết áp.Chúng ta sẽ tách nội dung của cột này ra làm 3 cột sau: PulseRate : giá trị huyết áp, Sex: giới tính ( m: male, f: female) và time: thời gian (tháng-ngày) như sau:

```python

#Melt the Sex + time range columns in single column
F= pd.melt(F,id_vars=['Id' ,'Age' , 'Weight' ,'Firstname' , 'Lastname'], value_name="PulseRate", var_name="sex_and_time").sort_values(['Id','Age' ,'Weight' ,'Firstname' ,'Lastname'])
# Tách dữ liệu từ cột "m0006" với biểu thức chính quy
tmp_df = F["sex_and_time"].str.extract(r"([mf])(\d{2})(\d{2})", expand=True)

# Đặt tên cho các cột mới
tmp_df.columns = ["Sex", "month", "day"]

# Tạo cột PulseRate từ giá trị huyết áp
F["PulseRate"] = F["sex_and_time"].str.extract(r"(\d+)$")  # Giả sử giá trị huyết áp là số ở cuối chuỗi

# Tạo cột Time dựa trên "month" và "day"
tmp_df["Time"] = tmp_df["month"] + "-" + tmp_df["day"]

# Gộp dữ liệu
F = pd.concat([F, tmp_df], axis=1)

# Xóa cột "m0006" nếu không cần thiết
F = F.drop(['sex_and_time'], axis=1)

F
```

Hãy khảo sát tỉ lệ dữ liệu thiếu trên biến huyết áp. Dữ liệu bị thiếu thì hãy xử lý bằng 
phương pháp sau
• Thay thế bằng giá trị trung bình liền trước và liền sau của người đó. Nếu không 
được thì dùng 2)
• Thay thế bằng giá trị trung bình 2 giá liền trước của người đó. Nếu không được 
thì dùng 3)
• Thay thế bằng giá trị trung bình 2 giá liền sau của người đó. Nếu không được thì 
dùng 4)
• Trung bình của các giá trị huyết áp của người đó. Nếu không được thì dùng 5).
• Trung bình của các giá trị huyết áp của nhóm giới tính. Nếu không được thì dùng 
6)
• Trung bình của các giá trị dữ liệu. Nếu không được thì thay bằng mức ổn định 
trong y học.

```python

# Chuyển đổi cột "PulseRate" thành kiểu số
F['PulseRate'] = pd.to_numeric(F['PulseRate'], errors='coerce')

#Khảo sát tỉ lệ dữ liệu thiếu trên biến huyết áp
missing_pulse_rate = F['PulseRate'].isnull().mean()
print(f"Tỉ lệ dữ liệu thiếu trên biến huyết áp: {missing_pulse_rate:.2%}")

# Hàm thay thế giá trị thiếu
def fill_missing_pulse_rate(row):
    if pd.isnull(row['PulseRate']):
        # Thay thế bằng giá trị trung bình liền trước và liền sau
        prev_value = F.loc[row.name - 1, 'PulseRate'] if row.name > 0 else None
        next_value = F.loc[row.name + 1, 'PulseRate'] if row.name < len(F) - 1 else None
        if prev_value is not None and next_value is not None:
            return (prev_value + next_value) / 2
        
        # Thay thế bằng giá trị trung bình 2 giá liền trước
        if row.name > 1:
            prev_values = F.loc[row.name - 2:row.name - 1, 'PulseRate']
            if not prev_values.isnull().any():
                return prev_values.mean()
        
        # Thay thế bằng giá trị trung bình 2 giá liền sau
        if row.name < len(F) - 2:
            next_values = F.loc[row.name + 1:row.name + 2, 'PulseRate']
            if not next_values.isnull().any():
                return next_values.mean()
        
        # Trung bình của các giá trị huyết áp của người đó
        person_values = F[(F['Firstname'] == row['Firstname']) & (F['Lastname'] == row['Lastname'])]['PulseRate']
        if not person_values.isnull().all():
            return person_values.mean()
        
        # Trung bình của các giá trị huyết áp của nhóm giới tính
        gender_values = F[F['Sex'] == row['Sex']]['PulseRate']
        if not gender_values.isnull().all():
            return gender_values.mean()
        
        # Trung bình của các giá trị dữ liệu
        return F['PulseRate'].mean()
    
    return row['PulseRate']

# Áp dụng hàm để thay thế giá trị thiếu
F['PulseRate'] = F.apply(fill_missing_pulse_rate, axis=1, result_type='reduce')
F
```
