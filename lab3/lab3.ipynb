```python
# đọc file
import pandas as pd
from pandas import ExcelFile

pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
pd.set_option('display.width', None)        

data='D:/Năm 4/kỳ 2/học sau/lab3/dirty_data.csv'
F=pd.read_csv(data)
F
```

 1. Kiểm tra dữ liệu bị thiếu:
     - Kiểm tra các cột có dữ liệu thiếu
     - Điền giá trị phù hợp hoặc loại bỏ các hàng có quá nhiều dữ liệu thiếu

```python
# kiểm tra các cột có dữ liệu thiếu 
F.isnull().sum()
```

```python
# điển giá trị phù hợp hoặc loại bỏ các hàng có nhiều dữ liệu thiếu
F.fillna("unknow", inplace=True)
F
```

```python
F.drop_duplicates(inplace=True)
F
```

2. Phát hiện và sửa lỗi dữ liệu bất thường:
    - Tìm các giá trị bất thường trong cột tuổi(age) có thể là âm hoặc quá lớn
    - Kiểm tra cột thu nhập (income) để phát hiện dữ liệu bị sai

```python

# Chuyển đổi cột 'age' và 'income' thành số, đặt các giá trị không thể chuyển đổi thành NaN
F['age'] = pd.to_numeric(F['age'], errors='coerce')
F['income'] = pd.to_numeric(F['income'], errors='coerce')

# Phát hiện các giá trị bất thường trong cột tuổi (age)
unusual_ages = F[(F['age'] < 0) | (F['age'] > 100)]

print("- Các giá trị tuổi bất thường:")
print(unusual_ages)



# Sửa các giá trị bất thường trong cột tuổi (age)
F.loc[(F['age'] < 0) | (F['age'] > 100), 'age'] = F['age'].mean()


# Kiểm tra các giá trị bất thường trong cột thu nhập (income)
unusual_income = F[F['income'] < 0]

print("- Các giá trị thu nhập bất thường:")
print(unusual_income)



# Sửa các giá trị bất thường trong cột thu nhập (income)
F.loc[F['income'] < 0, 'income'] = F['income'].mean()
print("đã sửa")
F

```

3. Chuẩn hoá dạng dữ liệu:
    - Chuyển đổi cột ngày sinh về dạng chuẩn yyyy-mm-dd
    - Chuẩn hoá dữ liệu trong cột giới tính (gender) để chỉ có hai giới tính hợp lệ là "Male" và "Female"

```python

# Chuyển đổi cột ngày sinh về dạng chuẩn yyyy-mm-dd

# Chuyển đổi định dạng ngày từ yyyy/mm/dd sang yyyy-mm-dd
F['dob'] = F['dob'].str.replace('/', '-')

F['dob'] = pd.to_datetime(F['dob'], errors='coerce').dt.strftime('%Y-%m-%d')

# Chuẩn hoá dữ liệu trong cột giới tính (gender)
def normalize_gender(gender):
    if gender.lower() in ['male', 'm']:
        return 'Male'
    elif gender.lower() in ['female', 'f']:
        return 'Female'
    else:
        return 'Unknown'  # Hoặc bạn có thể chọn loại bỏ các giá trị không hợp lệ

F['gender'] = F['gender'].apply(normalize_gender)
F
```

4. Xoá bỏ dữ liệu trùng lặp:
    - Xác định và xoá các bảng ghi trủng lặp đưa trên thông tin khách hàng

```python


# Xoá bỏ dữ liệu trùng lặp dựa trên thông tin khách hàng
# Giả sử thông tin khách hàng bao gồm các cột như 'customer_id', 'name', 'dob', v.v.
#điều chỉnh danh sách cột theo ý

F.drop_duplicates(subset=['customer_id', 'name', 'dob', 'email'], inplace=True)

# Kiểm tra dữ liệu sau khi xoá bỏ trùng lặp
print("Dữ liệu sau khi xoá bỏ trùng lặp:")
F

```
