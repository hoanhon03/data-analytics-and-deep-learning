14. Biết công thức tính chỉ số BMI như sau: 
 
BMI = Weight(kg)/Height2 (m2)   
 
Tạo một cột mới trong Dataframe với tên <cs_BMI>, tính chỉ số BIM mỗi người

```python
df['cs_BMI'] = (df['Weight'] / df['Height']**2)
print(df)
```

15. Theo thang đo kết quả tính BMI của châu Á: 
Chỉ số BMI dưới 18,5 là thiếu cân, 18.5-22,9 là bình   thường, 23-24,9 là thừa 
cân, từ 25 trở lên là béo phì. 
Tạo cột với tên <Phan_Loai> với các mức ‘0’,’1’,’2’,’3’ tương ứng với mức chỉ 
số BMI trên. 

```python
def bmi_status(cs_BMI):
    if cs_BMI < 18.5:
        return 0
    elif 18.5 <= cs_BMI < 24.9:
       return 1
    else:
       return 2

df['phan_loai'] = df['cs_BMI'].apply(bmi_status)
print(df)
```

16. Chuẩn hóa giá trị của cột cs_BMI chuẩn hóa giá trị trong đoạn từ 0 tới 1.

```python
# Chuẩn hóa cột 'cs_BMI' về khoảng giá trị từ 0 đến 1
df['BMI_normalized'] = (df['cs_BMI'] - df['cs_BMI'].min()) / (df['cs_BMI'].max() - df['cs_BMI'].min())

print(df)
```

17. Cho biết: 
   + ‘Mr‘tương ứng với ‘male’, 
+ ’Mrs’ và ‘Miss’ tương ứng với ‘female’.   
Tạo cột với tên <GT>, Sau khi thực hiện, xóa các từ Mr, Mrs, Miss nếu trong 
cột <Name> còn các từ đó.

```python
# Hàm extract_title nhận vào một DataFrame (df)
def extract_title(df):
    # Sử dụng regex để trích xuất các tiền tố trong cột 'secondName' như Mr, Mrs, Miss, Master
    # (\b ... \b) đảm bảo khớp chính xác với từ cần tìm, tránh nhầm lẫn với các từ có ký tự liền trước/sau
    # expand=False để trả về một Series thay vì DataFrame
    df['GT'] = df['Name'].str.extract(r'(\bMr\b|\bMrs\b|\bMiss\b)', expand=False)
    return df

# Gọi hàm extract_title để trích xuất tiền tố và thêm cột 'namePrefix' vào df
df = extract_title(df)

# Thông báo rằng đã chuyển đổi thành công sang các tiền tố như Mr, Mrs, Miss, Master
print('đã chuyển đổi sang mr,ms')
df
```

chuyển mr->m/mrs,miss-> fm

```python

def substitute_Sex(x_df):
    mapping={'Mr':'male', 'Mrs':'female', 'Miss':'female'}
    x_df['GT']=x_df['GT'].map(mapping)
    return x_df

tidisater5=df.pipe(substitute_Sex)
tidisater5
```
