```python
# bước 1:
# Lập bảng tần số và tần suất cho biến GT
gt_counts = F["GT"].value_counts()
gt_frequencies = F["GT"].value_counts(normalize=True)*100
# tạo bảng để hiện thị 
gt_table = pd.DataFrame({
    "Tần số": gt_counts,
    "Tần suất (%)": gt_frequencies
})
print("Bảng tần số và tần suất cho biến GT:")
print(gt_table)
# Vẽ biểu đồ tần số (cột)
plt.figure(figsize=(6, 4))
gt_counts.plot(kind="bar", color=["blue", "pink"])
plt.title("Biểu đồ tần số giới tính")
plt.xlabel("Giới tính")
plt.ylabel("Tần số")
plt.xticks(rotation=0)
plt.show()

# Vẽ biểu đồ tần suất (tròn)
plt.figure(figsize=(6, 6))
gt_counts.plot(kind="pie", autopct="%1.1f%%", colors=["blue", "pink"])
plt.title("Biểu đồ tần suất giới tính")
plt.ylabel("")
plt.show()
```

```python
# Bước 2: Hãy trình bày dữ liệu lần lượt các biến: US_TBM1, US_TBM2 và US_TBM3
# Hàm để lập bảng tần số và tần suất cho một biến
def create_frequency_table(column_name):
    counts = F[column_name].value_counts()
    frequencies = F[column_name].value_counts(normalize=True) * 100
    frequency_table = pd.DataFrame({'Tần số': counts, 'Tần suất (%)': frequencies})
    print(f"Bảng tần số và tần suất cho biến {column_name}:")
    print(frequency_table)
    print("\n")

# Trình bày dữ liệu cho US_TBM1
create_frequency_table('US_TBM1')

# Trình bày dữ liệu cho US_TBM2
create_frequency_table('US_TBM2')

# Trình bày dữ liệu cho US_TBM3
create_frequency_table('US_TBM3')

```

```python
# bước 3 trình bày dữ liệu biến DT với học sinh là nam (Male=M / Female= F) :

# Lọc dữ liệu cho học sinh nam
male_students = F[F['GT'] == 'M']

# Lập bảng tần số và tần suất cho biến DT
dt_counts = male_students["DT"].value_counts()
dt_frequencies = male_students["DT"].value_counts(normalize=True) * 100

# Tạo DataFrame để hiển thị tần số và tần suất
dt_table = pd.DataFrame({'Tần số': dt_counts, 'Tần suất (%)': dt_frequencies})
print("Bảng tần số và tần suất cho biến DT (học sinh nam):")
print(dt_table)
```

```python
# Bước 4: Hãy trình bày dữ liệu biến KV với các học sinh là nam thuộc dân tộc Kinh, có điểm thỏa 
# mãn điều kiện (DH1 >= 5.0 và DH2 >= 4.0 và DH3 >= 4.0)

# Lọc dữ liệu cho học sinh nam thuộc dân tộc Kinh và thỏa mãn điều kiện điểm
filtered_students = F[(F['GT'] == 'M') & (F['DT'] == 0.0) & (F['DH1'] >= 5.0) & (F['DH2'] >= 4.0) & (F['DH3'] >= 4.0)]

print(filtered_students)
# Lập bảng tần số và tần suất cho biến KV
kv_counts = filtered_students["KV"].value_counts()
kv_frequencies = filtered_students["KV"].value_counts(normalize=True) * 100

# Tạo DataFrame để hiển thị tần số và tần suất
kv_table = pd.DataFrame({'Tần số': kv_counts, 'Tần suất (%)': kv_frequencies})
print("Bảng tần số và tần suất cho biến KV (học sinh nam, dân tộc Kinh, điểm thỏa mãn):")
kv_table
```

```python
# bước 5: Hãy trình bày dữ liệu lần lượt các biến 
# DH1, DH2, DH3 lớn hơn bằng 5.0 và thuộc khu vực 2NT

# Lọc dữ liệu cho học sinh có điểm DH1, DH2, DH3 >= 5.0 và thuộc khu vực 2NT
filtered_students_2NT = F[(F['DH1'] >= 5.0) & (F['DH2'] >= 5.0) & (F['DH3'] >= 5.0) & (F['KV'] == '2NT')]

# Lập bảng tần số và tần suất cho các biến DH1, DH2, DH3
for column in ['DH1', 'DH2', 'DH3']:
    counts = filtered_students_2NT[column].value_counts()
    frequencies = filtered_students_2NT[column].value_counts(normalize=True) * 100
    table = pd.DataFrame({'Tần số': counts, 'Tần suất (%)': frequencies})
    print(f"Bảng tần số và tần suất cho biến {column} (điểm >= 5.0, khu vực 2NT):")
    print(table)
    print("\n")

    
## chỉ lấy khi thỏa mảng cả b điều kiện là DH1>=5/DH2>=5/DH3>=5##
```

```python

#2
# Tạo biến phân loại cho môn Toán (T1)
def classify_t1(score):
    if score < 5:
        return 'y'
    elif 5 <= score < 7:
        return 'tb'
    elif 7 <= score < 8:
        return 'k'
    else:
        return 'g'

F['phanloait1'] = F['T1'].apply(classify_t1)

#3
# Lập bảng tần số cho biến phanloait1
phanloait1_counts = F['phanloait1'].value_counts()
print("Bảng tần số cho biến phanloait1:")
print(phanloait1_counts)


#4 
# Vẽ biểu đồ đường Multiple Line cho biến T1 được phân loại bởi phanloait1
plt.figure(figsize=(10, 6))
for category in ['y', 'tb', 'k', 'g']:
    subset = F[F['phanloait1'] == category]
    plt.plot(subset.index, subset['T1'], linestyle='-', label=category)

plt.title('Biểu đồ đường Multiple Line cho biến T1 phân loại bởi phanloait1')
plt.xlabel('Chỉ số')
plt.ylabel('Giá trị T1')
plt.legend(title='Phân loại T1')
plt.grid(True)
plt.show()
```

```python
#5

# Vẽ biểu đồ Drop-line cho biến T1 được phân loại bởi phanloait1
plt.figure(figsize=(17, 9))
for category in ['k', 'tb', 'k', 'g']:
    subset = F[F['phanloait1'] == category]
    plt.stem(subset.index, subset['T1'], linefmt='-', markerfmt='o', basefmt=' ', label=category)

plt.title('Biểu đồ Drop-line cho biến T1 phân loại bởi phanloait1')
plt.xlabel('Chỉ số')
plt.ylabel('Giá trị T1')
plt.legend(title='Phân loại T1')
plt.grid(True)
plt.show()

```

```python
# Phần 5: Mô tả dữ liệu và khảo sát dạng phân phối
# khai báo 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
# 1. Mô tả và khảo sát phân phối cho biến T1
t1_description = F['T1'].describe()
print("Mô tả các thông số thống kê cơ bản của biến T1:")
print(t1_description)

# Vẽ Box-Plot cho biến T1
plt.figure(figsize=(10, 6))
sns.boxplot(x=F['T1'])
plt.title('Box-Plot của biến T1')
plt.xlabel('T1')
plt.show()

# Tính các đại lượng hướng tâm
mean_T1 = F['T1'].mean()
median_T1 = F['T1'].median()
mode_T1 = F['T1'].mode()[0]
print("Mean của biến T1:", mean_T1)
print("Median của biến T1:", median_T1)
print("Mode của biến T1:", mode_T1)

# Vẽ Histogram cho biến T1
plt.figure(figsize=(10, 6))
sns.histplot(F['T1'], bins=30, kde=True)
plt.title('Histogram của biến T1')
plt.xlabel('T1')
plt.ylabel('Tần suất')
plt.show()

# Tính mức độ lệch và mức độ nhọn
skewness_T1 = F['T1'].skew()
kurtosis_T1 = F['T1'].kurtosis()
print("Mức độ lệch của biến T1:", skewness_T1)
print("Mức độ nhọn của biến T1:", kurtosis_T1)

# Vẽ QQ-Plot cho biến T1
plt.figure(figsize=(10, 6))
stats.probplot(F['T1'], dist="norm", plot=plt)
plt.title('QQ-Plot của biến T1')
plt.show()

# 2. Mô tả và khảo sát phân phối cho biến T1 trên từng nhóm phân lớp (phanlopT1)
F['phanlopT1'] = pd.cut(F['T1'], bins=[0, 5, 7, 8, 10], labels=['kém', 'trung bình', 'khá', 'giỏi'])

# Box-Plot cho biến T1 theo phân lớp phanlopT1
plt.figure(figsize=(10, 6))
sns.boxplot(x='phanlopT1', y='T1', data=F)
plt.title('Box-Plot của biến T1 theo phân lớp phanlopT1')
plt.xlabel('Phân lớp T1')
plt.ylabel('T1')
plt.show()

# Histogram cho biến T1 theo phân lớp phanlopT1
g = sns.FacetGrid(F, col='phanlopT1', col_wrap=2, height=6)
g.map(sns.histplot, 'T1', bins=30, kde=True)
g.fig.suptitle('Histogram của biến T1 theo phân lớp phanlopT1', y=1.02)
plt.show()

# QQ-Plot cho biến T1 theo phân lớp phanlopT1
for label in F['phanlopT1'].unique():
    plt.figure(figsize=(10, 6))
    subset = F[F['phanlopT1'] == label]
    stats.probplot(subset['T1'], dist="norm", plot=plt)
    plt.title(f'QQ-Plot của biến T1 - Phân lớp {label}')
    plt.show()

# 3. Khảo sát tương quan giữa biến DH1 theo biến T1
correlation_DH1_T1 = F['DH1'].corr(F['T1'])
print("Hệ số tương quan giữa DH1 và T1:", correlation_DH1_T1)

# Scatter Plot giữa DH1 và T1
plt.figure(figsize=(10, 6))
sns.scatterplot(x='T1', y='DH1', data=F)
plt.title('Scatter Plot giữa DH1 và T1')
plt.xlabel('T1')
plt.ylabel('DH1')
plt.show()

# 4. Khảo sát tương quan giữa biến DH1 theo biến T1 trên từng nhóm khu vực
g = sns.FacetGrid(F, col='KV', col_wrap=3, height=6)
g.map(sns.scatterplot, 'T1', 'DH1')
g.fig.suptitle('Scatter Plot giữa DH1 và T1 theo từng nhóm khu vực', y=1.02)
plt.show()

# 5. Khảo sát tương quan giữa các biến DH1, DH2, DH3
correlation_matrix = F[['DH1', 'DH2', 'DH3']].corr()
print("Ma trận tương quan giữa DH1, DH2 và DH3:")
print(correlation_matrix)

# Scatter Plot giữa các biến DH1, DH2 và DH3
sns.pairplot(F[['DH1', 'DH2', 'DH3']])
plt.suptitle('Scatter Plot giữa các biến DH1, DH2 và DH3', y=1.02)
plt.show()

```
